```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def fx(x):
    return 5*x**2 - 2*x + 6

def dx(x):
    return 10*x - 2
```

```python
x = np.linspace(-2,2,2000)

plt.plot(x,fx(x),x,dx(x))
plt.xlim(-2,2)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["f(x)","df(x)"])
plt.grid()
plt.show()
```

```python
localmin = -5.0
learning_rate = 0.01
training_epochs = 100

for i in range(training_epochs):
    grad = dx(localmin)
    localmin = localmin - grad*learning_rate
localmin
```

```python
plt.plot(x,fx(x),x,dx(x))
plt.plot(localmin,dx(localmin),"ro")
plt.plot(localmin,fx(localmin),"ro")

plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["f(x)","df","f(x) min"])
plt.title('the emprical minimum: %s'%localmin[0])
plt.show()
```

```python
localmin = np.random.choice(x,1)
learning_rate = .01
training_epochs = 100

modelparams = np.zeros((training_epochs,2))

for i in range(training_epochs):
    grad = dx(localmin)
    localmin = localmin - grad*learning_rate
    modelparams[i,0] = localmin
    modelparams[i,1] = grad


fig,ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
    ax[i].plot(modelparams[:,i],"-o")
    ax[i].set_xlabel("iteration")
    ax[i].set_title(f'final minimum: {localmin[0]:.5f}')

ax[0].set_ylabel("local minimum")
ax[1].set_ylabel("Gradient")

plt.show()

```

```python
def Fx(x):
    return np.cos(2*np.pi*x) + x**2

def dF(x):
    return -2*np.pi*np.sin(2*np.pi*x) + 2*x

x = np.linspace(-2,2,2001)

plt.plot(x,Fx(x),x,dF(x))
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid()
plt.show()
```

```python
localmin = 0 #df/dx = 0 aka vanishing gradient
lr = .01
epoch = 100

for i in range(epoch):
    grad = dF(localmin)
    localmin = localmin - grad*lr

plt.plot(x,Fx(x),x,dF(x))
plt.plot(localmin,Fx(localmin),"ko")
plt.plot(localmin,dF(localmin),"ko")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["x","f(x)","f(x) min"])
plt.grid()
plt.show()
```
